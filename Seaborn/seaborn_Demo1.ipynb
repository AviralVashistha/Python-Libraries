# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
var=[1,2,3,4,5,6,7]
var_1=[2,3,4,1,5,6,7]
```

```python
plt.plot(var,var_1)
plt.show()
```

```python
x_1=pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var",y="var_1",data=x_1)
plt.show()
```

```python
sns.lineplot(x=var,y=var_1)
plt.show()
```

```python
y_1=sns.load_dataset("penguins")
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data=y_1)
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1, hue="sex",style="sex",palette="Accent",markers=['o','>'])
plt.show()
```

```python
y_1=sns.load_dataset("penguins").head(20)
y_1
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1, hue="sex",style="sex",palette="Accent",markers=['o','>'],dashes=False,legend="full")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=y_1, hue="sex",style="sex",palette="Accent",markers=['o','>'],dashes=False,legend="full")
plt.grid()
plt.title("penguins")
plt.show()
```

```python
sns.barplot(x=var.bill_depth_mm,y=var.bill_length_mm,color='g')
plt.show()
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm,hue=var.sex,hue_order=["Female","Male"],saturation=1,err_kws={'color': 'y','linewidth':2},capsize=0.9)
plt.show()
```

```python
sns.heatmap(x)
plt.show()
```

```python
var1=np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=True)
plt.show()
```

```python
ar=np.array([['a0','a1','a2','a3','a4'],
            ['b0','b1','b2','b3','b4']],dtype='<U2')
```

```python
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt='s')
```

```python
y={"fontsize":20,"color":"r"}
sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y,linewidth=4,linecolor="y")
plt.show()
```

```python
y={"fontsize":20,"color":"r"}
v=sns.heatmap(var1,vmin=0,vmax=10,cmap='PuOr',annot=True,annot_kws=y, cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel="Python",ylabel="Language")
sns.set(font_scale=50)
plt.show()
```

# Multiple Plot (Facet-Grid)

```python
var=sns.load_dataset("tips")
var
```

```python
fg=sns.FacetGrid(var)
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex")
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex", hue="day")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day", hue="sex")
fg.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="day", hue="sex")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex", hue="day")
fg.map(plt.bar,"total_bill","tip").add_legend()
plt.show()
```

```python
fg=sns.FacetGrid(var,col="sex", hue="day")
fg.map(plt.bar,"total_bill","tip", edgecolor="green").add_legend()
plt.show()
```
