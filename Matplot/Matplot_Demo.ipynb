```python
a=np.arange(5)
b=[2,4,6,8,10]
c=[5,6,7,8,9]
```

```python
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'k:',label='Periods')
ax.legend()
plt.title("Frequency of a Signal")
```

```python
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'k:',label='Periods')
ax.legend(loc = 'upper center')
plt.title("Frequency of a Signal")
```

```python
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'k:',label='Periods')
legend=ax.legend(loc = 'upper center')
plt.title("Frequency of a Signal")
legend.get_frame().set_facecolor('yellow')
```

```python
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'k:',label='Periods')
legend=ax.legend(loc = 'upper center', fontsize='x-large')
plt.title("Frequency of a Signal")
legend.get_frame().set_facecolor('yellow')
```

```python
student=np.array(["Tim","John","Amit","Jacob","Karl","Gary"])
marks=np.array([85,59,99,77,67,95])
```

```python
plt.bar(student,marks)
plt.xlabel("Student")
plt.ylabel("Marks")
plt.title("Student Marks")
```

```python
cricketer=np.array(["Tim","John","Amit","Jacob","Karl","Gary"])
runs=np.array([85,59,99,77,67,95])
```

```python
plt.pie(runs,labels=cricketer,autopct='%1.2f')
plt.title("Runs Scored in a Match")
```
