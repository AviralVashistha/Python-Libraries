# Problem Statement 12

**Imagine you are working for an e-commerce company, and the marketing team wants to target customers more effectively by creating segments based on their purchasing behaviour. However, traditional segmentation methods are not providing satisfactory results.**

Solution: Use DBSCAN (density based spatial clustering of application with noise) to identify clusters of customers with similar purchasing patterns.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data=np.random.rand(100,2)
```

```python
scaler=StandardScaler()
data_scaled=scaler.fit_transform(data)
```

```python
dbscan=DBSCAN(eps=0.3,min_samples=5)
clusters=dbscan.fit_predict(data_scaled)
```

```python
plt.scatter(data[:,0],data[:,1],c=clusters,cmap='viridis',marker='o',s=25)
plt.title("DBSCAN - Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
result_df=pd.DataFrame({'Feature 1':data[:,0],'Feature 2':data[:,1],'Cluster':clusters})
```

```python
print ("Numberof Clusters: ",len(np.unique(clusters)))
print("Size of each Cluster: ")
print(result_df['Cluster'].value_counts())
```

```python
user_input=np.array([[0.6,0.7]])
user_input_scanned=scaler.transform(user_input)
user_cluster=dbscan.fit_predict(user_input_scanned)
print("User Input belongs to cluster:",user_cluster[0])
```
