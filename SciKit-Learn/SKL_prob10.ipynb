# Problem Statement 10

**A mall wants to understand its customer base better to improve its marketing strategy. The goal is to cluster customers into roups based on their annual income and spending score, so the marketing team can tailor their campaigns to each cluster's preferences and behaviours.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(42)
data={'AnnualIncome': np.random.randint(30000,100000,100),'SpendingScore':np.random.randint(1,100,100)}
df=pd.DataFrame(data)
print(df)
```

```python
plt.scatter(df['AnnualIncome'],df['SpendingScore'])
plt.title("Customer Data- Annualincome vs Spending Score")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
X=df.values
kmeans=KMeans(n_clusters=3,random_state=42)
df['Cluster']=kmeans.fit_predict(X)
```

```python
print(df)
```

```python
plt.scatter(df['AnnualIncome'],df['SpendingScore'],c=df['Cluster'],cmap='rainbow')
plt.title("KMeans Clustering - Annualincome vs Spending Score")
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
user_input={'Annual Income':[75000],'SpendingScore':[70]}
user_df=pd.DataFrame(user_input)
user_cluster=kmeans.predict(user_df)
print(f"The user belongs to cluster {user_cluster[0]}")
```
